#Create a new column Classification that display "price(PHP)" if the value of outcome is price(USD)

```python
usd_to_php = 55
p_data['price(PHP)'] = p_data['price(USD)'] * usd_to_php
p_data
```

#Create a new dataframe "HighRam" that gathers data with ram(GB) greater than 16

#Display all the mid range phones available in the data set in php


```python
p_data = pd.DataFrame(p_data)

midrange = 25000
f = p_data[p_data['price(PHP)'] <= midrange]
print("price(PHP)\tphone_name")
for i, row in f.iterrows():
    print("₱{:.0f}\t\t{}".format(row['price(PHP)'], row['phone_name']))
```

#Display all the high end phone

```python
dfm = pd.DataFrame(p_data)

midrange = 25000
highend = dfm[dfm['price(PHP)'] >= midrange]
f = highend
print("price(PHP)\tphone_name")
for i, row in f.iterrows():
    print("₱{:.0f}\t\t{}".format(row['price(PHP)'], row['phone_name']))
```

#Display a specific brand

#Use numpy to get the average of inch and prices of phones

```python
m_usd = np.mean(p_data['price(USD)'])
m_php = np.mean(p_data['price(PHP)'])
m_inch = np.mean(p_data['inches'])

print("Average prices of phone in USD: ${:.2f}".format(m_usd))
print("Average prices of phone in PHP: ₱{:.2f}".format(m_php))
print("Average inch of phone in the dataset: {:.2f}".format(m_inch))
```

#Use numpy to get the Median inches and prices of phone.

```python
m_usd = np.median(p_data['price(USD)'])
m_php = np.median(p_data['price(PHP)'])
m_inch = np.median(p_data['inches'])

print("Median prices of phone in USD: ${:.2f}".format(m_usd))
print("Median prices of phone in PHP: ₱{:.2f}".format(m_php))
print("Median inch of phone in the dataset: {:.2f}".format(m_inch))
```

#Use numpy to get the middle values of inches of phone.

```python
m_usd = np.median(p_data['price(USD)'])
m_php = np.median(p_data['price(PHP)'])
m_inch = np.median(p_data['inches'])

print("Median prices of phone in USD: ${:.2f}".format(m_usd))
print("Median prices of phone in PHP: ₱{:.2f}".format(m_php))
print("Median inch of phone in the dataset: {:.2f}".format(m_inch))
```

#Use numpy to get the standard deviation of the phone prices.

```python
std_usd = np.std(p_data['price(USD)'])
std_php = np.std(p_data['price(PHP)'])


print("standard deviation prices of phone in USD: ${:.2f}".format(std_usd))
print("standard deviation prices of phone in PHP: ₱{:.2f}".format(std_php))
```
