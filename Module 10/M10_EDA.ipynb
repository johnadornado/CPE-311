```python
# Set the ID as the df index
salary_df.set_index('ID', inplace=True)
```

```python
# Use a lambda function to change Gender from 'm' to 0 and 'f' to 1
salary_df.Gender = salary_df.Gender.apply(lambda x : 0 if x == 'm' else 1)
```

# Data Plotting
Tips:
- If you want to plot CATEGORICAL data use: barcharts and boxplots.
- If you want to plot NUMERICAL data use: histograms, scatterplot, line graphs, boxplots.

```python
sns.countplot(x = 'Gender', data=salary_df)
```

```python
sns.violinplot(x = 'Gender', y = 'Salary', data = salary_df)
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1, 2, 1)
sns.boxplot(x = 'Salary', data = salary_df)
plt.subplot(1, 2, 2)
sns.histplot(x = 'Salary', data = salary_df, kde = True)
```

```python
plt.figure(figsize=(8,8))
salary_df['Degree'].value_counts().plot(kind='pie', autopct='%1.0f%%')
```

```python
sns.catplot(x = 'Degree', y = 'Salary', data = salary_df)
```

```python
sns.barplot(x = 'CollegeTier', y = 'Salary', data = salary_df)
```
